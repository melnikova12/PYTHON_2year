```python
def get_posts():
    posts_massive = []
    offset = 0 
    while offset < 200:
        posts_raw = API.wall.get(domain=DOMAIN, count=100, offset=offset, filter=all)
        if not posts_raw['items']:
            break
        posts_massive.extend(posts_raw['items'])
        offset += 100
    with open(NAME_OF_TXT, 'w', encoding='utf-8') as f:
        for i in posts_massive:
            owner_id = i['owner_id']
            date = i['date']
            text = i['text']
            utc = datetime.fromtimestamp(date)
            f.write(str(owner_id) + '\n')
            f.write(utc.strftime('%a, %d-%m-%Y\n%H:%M:%S\n'))
            
            f.write(text)
            f.write('\n\n')
    return posts_massive
```

```python
def get_comments():
    comments_massive = []
    for i in posts:
        owner_id = i['owner_id']
        post_id = i['id']
        offset = 0
        temp = []
        while True:
            comments = API.wall.getComments(owner_id=owner_id, post_id=post_id, count=100, offset=offset)
            if not comments['items']:
                break
            comments_massive.extend(comments['items'])
            temp.extend(comments['items'])
            offset += 100
            sleep(1)
        POSTS_DICT[(i['id'], count_len(i))] = count_average_len(temp)
        
    with open(NAME_OF_TXT, 'a', encoding='utf-8') as f:
        for i in comments_massive:
            try:
                owner_id = i['from_id']
                date = i['date']
                text = i['text']
            except KeyError:
                continue
            utc = datetime.fromtimestamp(date)
            f.write(str(owner_id) + '\n')
            f.write(utc.strftime('%a, %d-%m-%Y\n%H:%M:%S\n'))
            
            f.write(text)
            f.write('\n\n')
    return comments_massive
```

```python
def count_len(post):
    try:
        text = post['text']
        text = text.split()
        return len(text)
    except KeyError:
        return 0

    
def count_average_len(posts):
    lens = [count_len(post) for post in posts]
    try: 
        return sum(lens) / len(lens)
    except ZeroDivisionError:
        return 0
```

```python
def get_users():
    users = []
    com_users = []
    users_packs = []
    for i in comments:
        try:
            # для того, чтобы отсеять комментарии от сообществ 
            if i['from_id'] > 0:
                com_users.append(i['from_id'])
        except KeyError:
            continue 
    i = 0 
    while i < len(com_users):
        if len(com_users) - i > 1000:
            users_packs.append(com_users[i:i+1000])
        else:
            users_packs.append(com_users[i:])
        i += 1000
    for i in users_packs:
        i = [str(j) for j in i]
        request = ','.join(i)
        users_raw = API.users.get(user_ids=request, fields='city,sex')
        users.extend(users_raw)
        sleep(1)
    return users
```

```python
def posts_data():
    post_summary = []
    for i in posts:
        post_len = count_len(i)
        date = datetime.fromtimestamp(i['date'])
        weekday = date.weekday()
        month = date.month
        post_summary.append((post_len, weekday, month))
    return post_summary


post_summary_data = posts_data()
```

```python
def graph_post_weekday():
    dict_post_weekday = {}
    for post in post_summary_data:
        if post[1] not in dict_post_weekday.keys():
            dict_post_weekday[post[1]] = [post[0]]
        else:
            dict_post_weekday[post[1]].append(post[0])
    for key in dict_post_weekday.keys():
        av = sum(dict_post_weekday[key]) / len(dict_post_weekday[key])
        dict_post_weekday[key] = av
            
    df = pd.DataFrame(dict(x=[i for i in dict_post_weekday.keys()], y=[i for i in dict_post_weekday.values()]))
    ax = sb.catplot('x','y', data=df, kind='bar', palette=sb.color_palette('husl',8))
    ax.set_axis_labels('День недели', 'Средняя длина поста')
    plt.show()
    

graph_post_weekday()    
```

```python
def comments_users_data():
    users_dict = {i['id']: i for i in users}
    comments_users = []
    for i in comments:
        try:
            len_com = count_len(i)
            user_id = i['from_id']
            if user_id < 0:
                continue
        except KeyError:
            continue
        try:
            sex = users_dict[user_id]['sex']
        except KeyError:
            sex = 0
        try:
            city = users_dict[user_id]['city']['title']
        except KeyError:
            city = 'unknown'
        comments_users.append((len_com, sex, city))
    return comments_users


comments_users = comments_users_data()
```

```python
def lemmatized_text():
    with open(NAME_OF_TXT, encoding='utf-8') as f:
        text = f.read()
    m = Mystem()
    lemmas = m.lemmatize(text)
    with open(NAME_OF_LEM_TXT, 'w', encoding='utf-8') as f:
        f.write(' '.join(lemmas))
```

```python
def frequency_count(name):
    with open(name, encoding='utf-8') as f:
        text = f.read()
    text = text.split()
    dict_freq = {}
    days_of_week = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
    for word in text:
        if word.isalpha() and (word not in days_of_week):
            if word not in dict_freq.keys():
                dict_freq[word] = text.count(word)
    dict_freq = [(dict_freq[i], i) for i in dict_freq.keys()]
    dict_freq.sort(reverse=True)
    
    return dict_freq[:20]


freq_plain = frequency_count(NAME_OF_TXT)
freq_lem = frequency_count(NAME_OF_LEM_TXT)
```

```python
def graph_frequency_plain():
    df = pd.DataFrame(dict(x=[i[1] for i in freq_plain], y=[i[0] for i in freq_plain]))
    ax = sb.catplot('y','x', data=df, kind='bar', palette=sb.color_palette('husl',8))
    ax.set_axis_labels('Частотность', 'Слово')
    plt.show()
    

graph_frequency_plain()
```

```python
def graph_frequency_lemmatized():
    df = pd.DataFrame(dict(x=[i[1] for i in freq_lem], y=[i[0] for i in freq_lem]))
    ax = sb.catplot('y','x', data=df, kind='bar', palette=sb.color_palette('husl',8))
    ax.set_axis_labels('Частотность', 'Лемма')
    plt.show()
    

graph_frequency_lemmatized()
```
